```python
import pandas as pd

df = pd.read_csv("learning.csv")
df
```

```python
df.plot(kind="scatter", x="toneladas", y="preco");
```

```python
from sklearn import linear_model

PImodelo= linear_model.LinearRegression()
x = df[["toneladas"]]
y = df[["preco"]]

PImodelo.fit(x, y)
```

```python
print(PImodelo.predict([[25]])[0][0])
```
